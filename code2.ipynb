Load the Book

```python
with open('miracle_in_the_andes.txt', 'r') as f:
    book = f.read()
```

Most used words

```python
import re
pattern = re.compile('[a-zA-Z]+')
re.findall(pattern, book)[:20]
```

```python
d = {}
for word in re.findall(pattern, book.lower()):
    if word in d:
        d[word] += 1
    else:
        d[word] = 1
```

```python
d_list = [(value, key) for (key, value) in d.items()]
sorted(d_list, reverse=True)
```

What are the most used words except articles and pronouns

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
english_stopwords = stopwords.words('english')
```

```python
filtered_words = []
for count, word in d_list:
    if word not in english_stopwords:
        filtered_words.append({word, count})
```

Sentiment Analysis : Most positive

```python
from nltk.sentiment import SentimentIntensityAnalyzer  # class
nltk.download('vader_lexicon')
```

Instantiate the class by creating its object instances

```python
analyzer = SentimentIntensityAnalyzer()
```

This object analyzer has its methods

```python
dir(analyzer)
```

```python
score = analyzer.polarity_scores("Hey, look how ugly the trees are, I hate them. I really hate them")
```

```python
if score["pos"] > score["neg"]:
    print("It is a Positive text")
else:
    print("It is a Negative text")
```

Entire Book sentiment analysis

```python
analyzer.polarity_scores(book)
```

Chapters sentiment analysis

```python
import re
pattern = re.compile('Chapter [0-9]+')
chapters = re.split(pattern, book)
```

```python
chapters[:1]
```

```python
len(chapters)
```

Excluding first empty string in chapters

```python
chapters = chapters[1:]
```

```python
for nr, chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    print(f"Chapter {nr+1} : {scores}")
```
